## Data Loading and Initial Preparation

```python
import os
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

df = pd.read_stata("./data/Cleandata.dta")


data = df.copy()
data.head()
```

```python
# Create initial assignment groups from fyclasstype
df['initial_small'] = (df['fyclasstype'] == 1).astype(int)
df['initial_regular'] = (df['fyclasstype'] == 2).astype(int)
df['initial_ra'] = (df['fyclasstype'] == 3).astype(int)

print(f"Initial assignment distribution:")
print(f"  Small: {df['initial_small'].sum()} ({df['initial_small'].mean()*100:.1f}%)")
print(f"  Regular: {df['initial_regular'].sum()} ({df['initial_regular'].mean()*100:.1f}%)")
print(f"  Regular/Aide: {df['initial_ra'].sum()} ({df['initial_ra'].mean()*100:.1f}%)")
```

```python
# ==========================================
# Prepare Demographic Variables
# ==========================================

# 1. Gender: Create a binary 'girl' indicator (1 = female, 0 = male/missing)
df['girl'] = df['gender'].str.lower().map({'male': 0, 'female': 1}).fillna(0).astype(int)

# 2. Race/Ethnicity: Ensure binary format for White/Asian indicator
df['white_asian'] = df['WhiteAsian'].fillna(0).astype(int)

# 3. Age: Calculate age as of September 1, 1985
# Convert day and year to numeric, coercing any errors to NaN
df['birth_day'] = pd.to_numeric(df['birthday'], errors='coerce')
df['birth_year'] = pd.to_numeric(df['birthyear'], errors='coerce')

# Map string months to numeric values
month_map = {
    'january': 1, 'february': 2, 'march': 3, 'april': 4,
    'may': 5, 'june': 6, 'july': 7, 'august': 8,
    'september': 9, 'october': 10, 'november': 11, 'december': 12
}

# Use a temporary series for the numeric month to avoid cluttering the DataFrame
birth_month_num = df['birthmonth'].str.lower().map(month_map)

# Calculate base age based purely on the birth year
df['age_1985'] = 1985 - df['birth_year']

# Adjust age for students born strictly after September 1st 
# (i.e., they haven't had their 1985 birthday yet by Sept 1)
born_after_sept_1st = (birth_month_num > 9) | ((birth_month_num == 9) & (df['birth_day'] > 1))
df.loc[born_after_sept_1st, 'age_1985'] -= 1

# Note: age_1985 will be a float due to NaN values (from errors='coerce'). 
# If you need it as an integer and are sure there are no NaNs, you can uncomment the line below:
# df['age_1985'] = df['age_1985'].astype(int)
```

```python
# Prepare free lunch variables
for grade in ['0', '1', '2', '3']:
    col = f'g{grade}freelunch_2'
    if col in df.columns:
        df[f'free_lunch_{grade}'] = df[col].fillna(0).astype(int)
```

```python
# Convert test scores to percentiles as described in the article
# First, ensure test scores are numeric
for grade in ['0', '1', '2', '3']:
    for subject in ['treadss', 'tmathss', 'wordskillss']:
        col = f'g{grade}{subject}'
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert class type columns to numeric categories
for grade in ['0', '1', '2', '3']:
    col = f'g{grade}classtype'
    if col in df.columns:
        # Map string values to numeric codes
        df[col] = df[col].astype(str)
        class_mapping = {
            'SMALL CLASS': 1,
            'REGULAR CLASS': 2, 
            'REGULAR + AIDE CLASS': 3,
            'REGULAR+AIDE CLASS': 3,
            'SMALL': 1,
            'REGULAR': 2,
            'REGULAR+AIDE': 3
        }
        df[col] = df[col].map(class_mapping)
```

```python
def calculate_percentile_score(row, grade, control_group_scores):
    """Calculate percentile rank relative to control group."""
    score = row
    if pd.isna(score):
        return np.nan
    # Percentile = (number of scores below this score / total scores) * 100
    return (np.sum(control_group_scores < score) / len(control_group_scores)) * 100
```

```python
# Calculate percentiles for each grade
for grade in ['0', '1', '2', '3']:
    # Get the average of three SAT tests
    test_cols = [f'g{grade}treadss', f'g{grade}tmathss', f'g{grade}wordskillss']
    if all(col in df.columns for col in test_cols):
        df[f'sat_avg_{grade}'] = df[test_cols].mean(axis=1)
        
        # Get control group (regular and regular/aide) for this grade
        class_col = f'g{grade}classtype'
        if class_col in df.columns:
            # Control group = regular (2) and regular/aide (3)
            control_mask = df[class_col].isin([2, 3])
            control_scores = df.loc[control_mask, f'sat_avg_{grade}'].dropna()
            
            if len(control_scores) > 0:
                # Calculate percentile for each student
                df[f'percentile_{grade}'] = df[f'sat_avg_{grade}'].apply(
                    lambda x: calculate_percentile_score(x, grade, control_scores)
                )
                
                # Verify: control group should have mean percentile around 50
                control_percentiles = df.loc[control_mask, f'percentile_{grade}'].mean()
                print(f"  Grade {grade}: Control group N={len(control_scores)}, Mean percentile={control_percentiles:.1f}")
```

```python
# Determine entry grade 
def get_entry_grade(row):
    """
    Determine the grade when student first entered STAR program.
    Based on initial assignment (fyclasstype) and when they actually have data.
    """
    # First check: which grade has class type data AND test scores?
    for grade in ['0', '1', '2', '3']:
        class_col = f'g{grade}classtype'
        test_col = f'g{grade}treadss'
        
        # Check if this student has data for this grade
        if class_col in row and pd.notna(row[class_col]):
            if test_col in row and pd.notna(row[test_col]):
                return grade
    return np.nan

df['entry_grade'] = df.apply(get_entry_grade, axis=1)

print(f"Entry grade distribution:")
entry_counts = df['entry_grade'].value_counts().sort_index()
for grade, count in entry_counts.items():
    grade_name = 'Kindergarten' if grade == '0' else f'Grade {grade}'
    print(f"  {grade_name}: {count} ({count/len(df)*100:.1f}%)")
```

## OLS Regression Analysis

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# ==========================================
# Regression Core Function
# ==========================================

def run_regression_extended(data, grade, model_type, controls_level):
    """
    Prepares the design matrix and runs OLS or Reduced Form regressions
    with varying levels of controls and school fixed effects.
    """
    # 1. Filter data and define target variables
    df_g = data[data['grade'] == grade].copy()
    
    if model_type == 'ols':
        treatment_vars = ['small_class', 'reg_aide_class']
    else:  # reduced_form
        treatment_vars = ['initial_small', 'initial_ra']
        
    student_controls = ['white_asian', 'girl', 'free_lunch']
    teacher_controls = ['teacher_white', 'teacher_exp', 'teacher_masters']
    
    # 2. Determine control variables based on specification level
    controls = []
    use_school_fe = False
    
    if controls_level >= 2:
        use_school_fe = True
    if controls_level >= 3:
        controls.extend(student_controls)
    if controls_level >= 4:
        controls.extend(teacher_controls)
        
    # 3. Build dataset for regression
    req_cols = ['percentile'] + treatment_vars + controls
    if use_school_fe:
        req_cols.append('school_id')
        
    # Keep only available columns and drop missing observations
    req_cols = [c for c in req_cols if c in df_g.columns]
    df_reg = df_g[req_cols].dropna().copy()
    
    if len(df_reg) < 10:
        return None

    # 4. Construct Design Matrix (X) and Outcome (y)
    y = df_reg['percentile'].astype(float)
    X = df_reg[treatment_vars + controls].astype(float)
    X = sm.add_constant(X)
    
    if use_school_fe:
        # Generate school fixed effects, dropping the first to avoid perfect multicollinearity
        school_dummies = pd.get_dummies(df_reg['school_id'].astype(str), prefix='sch', drop_first=True, dtype=float)
        X = pd.concat([X, school_dummies], axis=1)

    # 5. Fit Model with Fallback Standard Errors
    try:
        # Primary: Clustered standard errors at the school level
        model = sm.OLS(y, X).fit(cov_type='cluster', cov_kwds={'groups': df_reg['school_id']})
    except Exception:
        try:
            # Fallback 1: Heteroskedasticity-robust standard errors (HC1)
            model = sm.OLS(y, X).fit(cov_type='HC1')
        except Exception:
            # Fallback 2: Standard Homoskedastic OLS
            model = sm.OLS(y, X).fit()

    # 6. Return compiled results
    return {
        'params': model.params.to_dict(),
        'bse': model.bse.to_dict(),
        'r2': model.rsquared,
        'n_obs': len(df_reg),
        'school_fe': use_school_fe,
        'controls_level': controls_level,
        'model_type': model_type
    }

# ==========================================
# Formatting Helpers
# ==========================================

def format_coef_se(coef, se, add_stars=False):
    """Formats coefficient and standard error, optionally adding significance stars."""
    if coef is None or se is None or pd.isna(coef):
        return "—"
    
    stars = ""
    if add_stars and se > 0:
        t_stat = abs(coef / se)
        if t_stat > 2.576:   stars = "***" # p < 0.01
        elif t_stat > 1.960: stars = "**"  # p < 0.05
        elif t_stat > 1.645: stars = "*"   # p < 0.10
            
    # Format with 2 decimal places (3 if SE is very small)
    decimals = 3 if abs(se) < 0.01 else 2
    return f"{coef:.{decimals}f}{stars} ({se:.{decimals}f})"

# ==========================================
# Execute Regressions Across Grades
# ==========================================

results = {}
grades = [0, 1, 2, 3]
grade_names = ['Kindergarten', 'Grade 1', 'Grade 2', 'Grade 3']

for grade_num, grade_name in zip(grades, grade_names):
    grade_results = {}
    
    for col in [1, 2, 3, 4]:
        # Run OLS (Actual assignment)
        grade_results[f'ols_{col}'] = run_regression_extended(long_df, grade_num, 'ols', col)
        # Run Reduced Form (Initial assignment)
        grade_results[f'rf_{col}'] = run_regression_extended(long_df, grade_num, 'reduced_form', col)
        
    results[grade_name] = grade_results

# ==========================================
# Generate Summary Output
# ==========================================

print("\n" + "="*100)
print("SUMMARY OF MAIN RESULTS - Small Class Effects")
print("="*100)

summary_headers = ["Grade", "OLS (3)", "OLS (4)", "Reduced Form (7)", "Reduced Form (8)"]
print(" | ".join([h.ljust(18) for h in summary_headers]))
print("-" * 100)

for grade_name in grade_names:
    grade_res = results[grade_name]
    row_vals = [grade_name]
    
    for spec, var_name in [('ols_3', 'small_class'), ('ols_4', 'small_class'), 
                           ('rf_3', 'initial_small'), ('rf_4', 'initial_small')]:
        
        res = grade_res.get(spec)
        if res:
            c = res['params'].get(var_name)
            s = res['bse'].get(var_name)
            row_vals.append(format_coef_se(c, s, add_stars=True))
        else:
            row_vals.append("—")
            
    print(" | ".join([str(x).ljust(18) for x in row_vals]))
    
print("\nNote: *** p<0.01, ** p<0.05, * p<0.10")
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from linearmodels.iv import IV2SLS

# ==========================================
# Data Preparation
# ==========================================

# Create initial_regular variable (Regular without aide)
long_df['initial_regular'] = 1 - long_df['initial_small'] - long_df['initial_ra']

# Ensure logical consistency (clip any negative values to 0)
if (long_df['initial_regular'] < 0).any():
    long_df['initial_regular'] = long_df['initial_regular'].clip(lower=0)

# ==========================================
# 2SLS Regression Core Function
# ==========================================

def run_2sls_regression_fixed(data, grade, include_controls=True, include_school_fe=True):
    """
    Runs a 2SLS regression for a specific grade using the Matrix API.
    Returns correct 2SLS standard errors via the linearmodels package.
    """
    df_g = data[data['grade'] == grade].copy()
    
    # Define required variables
    # Note: Assuming 'small_class' is the endogenous treatment variable
    req_vars = ['percentile', 'small_class', 'initial_small', 'initial_regular']
    df_g = df_g.dropna(subset=req_vars)
    
    if len(df_g) < 10:
        return None

    # --------------------------------------------------------
    # 1. Define Exogenous Controls & Handle Missing Values
    # --------------------------------------------------------
    exog_cols = []
    
    if include_controls:
        # Impute missing continuous variables with median, binary/categorical with mean
        if 'teacher_exp' in df_g.columns:
            df_g['teacher_exp'] = df_g['teacher_exp'].fillna(df_g['teacher_exp'].median())
            
        student_controls = ['white_asian', 'girl', 'free_lunch']
        teacher_controls = ['teacher_white', 'teacher_exp', 'teacher_masters']
        
        for col in student_controls + teacher_controls:
            if col in df_g.columns:
                #df_g[col] = df_g[col].fillna(df_g[col].mean())
                df_g[col] = df_g[col].fillna(df_g[col].mode()[0])
                exog_cols.append(col)

    # Add School Fixed Effects
    if include_school_fe:
        # Create dummies and drop the first to avoid perfect multicollinearity (dummy trap)
        school_dummies = pd.get_dummies(df_g['school_id'].astype(str), prefix='sch', drop_first=True, dtype=float)
        df_g = pd.concat([df_g, school_dummies], axis=1)
        exog_cols.extend(school_dummies.columns)

    # --------------------------------------------------------
    # 2. Construct Regression Matrices
    # --------------------------------------------------------
    y = df_g['percentile']
    D = df_g[['small_class']] # Endogenous variable
    Z = df_g[['initial_small', 'initial_regular']] # Instruments
    
    # Exogenous matrix (Controls + FE + Constant)
    if exog_cols:
        X_exog = sm.add_constant(df_g[exog_cols].astype(float))
    else:
        # If no controls, X_exog is just a column of ones (the constant)
        X_exog = sm.add_constant(pd.DataFrame(np.ones(len(df_g)), index=df_g.index, columns=['const']))

    # --------------------------------------------------------
    # 3. First Stage (Extract F-Statistic)
    # --------------------------------------------------------
    # Regress endogenous variable on instruments + exogenous controls
    first_stage_exog = pd.concat([Z, X_exog], axis=1)
    fs_model = sm.OLS(D, first_stage_exog).fit()
    fs_f_stat = fs_model.fvalue

    # --------------------------------------------------------
    # 4. Standard OLS (For Baseline Comparison)
    # --------------------------------------------------------
    ols_exog = pd.concat([D, X_exog], axis=1)
    ols_model = sm.OLS(y, ols_exog).fit()
    
    # --------------------------------------------------------
    # 5. IV / 2SLS Estimation (linearmodels)
    # --------------------------------------------------------
    # We use linearmodels instead of manual 2-stage OLS to ensure standard errors
    # are mathematically adjusted for the first-stage estimation variance.
    try:
        iv_model = IV2SLS(dependent=y, exog=X_exog, endog=D, instruments=Z).fit(cov_type='robust')
    except Exception as e:
        print(f"  Grade {grade} IV2SLS Error: {e}")
        return None

    # Compile and return results
    return {
        'n_obs': len(df_g),
        'fs_f_stat': fs_f_stat,
        'ols_coef': ols_model.params['small_class'],
        'ols_se': ols_model.bse['small_class'],
        'iv_coef': iv_model.params['small_class'],
        'iv_se': iv_model.std_errors['small_class']
    }

# ==========================================
# Execute Regressions and Format Table VII
# ==========================================

grades = [0, 1, 2, 3]
grade_names = ['Kindergarten', 'Grade 1', 'Grade 2', 'Grade 3']
table_rows = [["Grade", "OLS", "2SLS", "Sample Size", "First Stage F"]]

for grade_num, grade_name in zip(grades, grade_names):
    res = run_2sls_regression_fixed(long_df, grade_num, include_controls=True, include_school_fe=True)
    
    if res:
        ols_str = f"{res['ols_coef']:.2f} ({res['ols_se']:.2f})"
        iv_str  = f"{res['iv_coef']:.2f} ({res['iv_se']:.2f})"
        n_obs   = f"{res['n_obs']:,}"
        fs_f    = f"{res['fs_f_stat']:.1f}"
        
        table_rows.append([grade_name, ols_str, iv_str, n_obs, fs_f])
    else:
        table_rows.append([grade_name, "—", "—", "—", "—"])

# Print Final Formatted Table
print("\n" + "="*100)
print("TABLE VII - OLS AND 2SLS ESTIMATES OF EFFECT OF SMALL CLASS ON ACHIEVEMENT")
print("DEPENDENT VARIABLE: AVERAGE PERCENTILE SCORE ON SAT")
print("="*100)

for row in table_rows:
    print(f"{row[0]:<15} {row[1]:<20} {row[2]:<20} {row[3]:<15} {row[4]:<15}")
print("-" * 90)
print("Note: All models include school fixed effects, student race, gender, free lunch status,")
print("teacher race, experience, and education. Standard errors in parentheses.")
```

## TSLS without covariates

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from linearmodels.iv import IV2SLS

# ==========================================
# Data Preparation
# ==========================================

# Ensure 'initial_regular' exists and is strictly non-negative
if 'initial_regular' not in long_df.columns:
    long_df['initial_regular'] = 1 - long_df['initial_small'] - long_df['initial_ra']
    long_df['initial_regular'] = long_df['initial_regular'].clip(lower=0)

# ==========================================
# Unconditional IV Regression Function
# ==========================================

def run_iv_no_covariates(data, grade):
    """
    Runs a baseline IV / 2SLS regression for a specific grade.
    This specification contains NO covariates and NO school fixed effects.
    Endogenous regressor: small_class
    Instrument: initial_small
    """
    # 1. Filter data and isolate required variables
    df_g = data[data['grade'] == grade].copy()
    req_vars = ['percentile', 'small_class', 'initial_small']
    df_g = df_g.dropna(subset=req_vars)
    
    if len(df_g) < 10:
        return None

    # 2. Define regression matrices
    y = df_g['percentile']
    D = df_g[['small_class']]     # Endogenous variable
    Z = df_g[['initial_small']]   # Instrument
    
    # Since there are no covariates, our exogenous matrix is just a constant (intercept)
    X_exog = sm.add_constant(pd.DataFrame(np.ones(len(df_g)), index=df_g.index, columns=['const']))

    # 3. First Stage (Extract F-Statistic for instrument strength)
    first_stage_exog = pd.concat([Z, X_exog], axis=1)
    fs_model = sm.OLS(D, first_stage_exog).fit()
    fs_f_stat = fs_model.fvalue

    # 4. Standard OLS (For baseline comparison, ignoring endogeneity)
    ols_exog = pd.concat([D, X_exog], axis=1)
    ols_model = sm.OLS(y, ols_exog).fit()

    # 5. IV / 2SLS Estimation (linearmodels)
    # This automatically computes the correctly adjusted second-stage standard errors.
    try:
        iv_model = IV2SLS(dependent=y, exog=X_exog, endog=D, instruments=Z).fit(cov_type='robust')
    except Exception as e:
        print(f"  Grade {grade} IV2SLS Error: {e}")
        return None

    # Compile and return results
    return {
        'n_obs': len(df_g),
        'fs_f_stat': fs_f_stat,
        'ols_coef': ols_model.params['small_class'],
        'ols_se': ols_model.bse['small_class'],
        'iv_coef': iv_model.params['small_class'],
        'iv_se': iv_model.std_errors['small_class']
    }

# ==========================================
# Execute Regressions Across Grades
# ==========================================

grades = [0, 1, 2, 3]
grade_names = ['Kindergarten', 'Grade 1', 'Grade 2', 'Grade 3']
table_rows = [["Grade", "OLS", "IV", "Sample Size", "First Stage F"]]

for grade_num, grade_name in zip(grades, grade_names):
    res = run_iv_no_covariates(long_df, grade_num)
    
    if res:
        ols_str = f"{res['ols_coef']:.2f} ({res['ols_se']:.2f})"
        iv_str  = f"{res['iv_coef']:.2f} ({res['iv_se']:.2f})"
        n_obs   = f"{res['n_obs']:,}"
        fs_f    = f"{res['fs_f_stat']:.1f}"
        
        table_rows.append([grade_name, ols_str, iv_str, n_obs, fs_f])
    else:
        table_rows.append([grade_name, "—", "—", "—", "—"])

# ==========================================
# Generate Formatted Output Table
# ==========================================

print("\n\n" + "="*100)
print("IV ESTIMATES OF EFFECT OF SMALL CLASS ON ACHIEVEMENT (NO COVARIATES)")
print("DEPENDENT VARIABLE: AVERAGE PERCENTILE SCORE ON SAT")
print("="*100)

for row in table_rows:
    print(f"{row[0]:<15} {row[1]:<20} {row[2]:<20} {row[3]:<15} {row[4]:<15}")
print("-" * 90)

print("Note: No covariates and no school fixed effects are included.")
print("Robust standard errors are shown in parentheses.")
```

## DDML LATE

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import doubleml as dml

def calculate_late_ddml(data, grade):
    df_g = data[data['grade'] == grade].copy()
    
    # Define covariates (matching your previous lists)
    covariates = [
        'white_asian', 'girl', 'free_lunch', 
        'teacher_white', 'teacher_exp', 'teacher_masters'
    ]
    # Add school dummies to covariates
    school_dummies = pd.get_dummies(df_g['school_id'].astype(str), prefix='sch', drop_first=True)
    df_g = pd.concat([df_g, school_dummies], axis=1)
    x_cols = covariates + list(school_dummies.columns)
    
    # Drop NAs
    df_g = df_g.dropna(subset=['percentile', 'small_class', 'initial_small'] + covariates)
    
    # Setup DoubleML Data Object
    # Y = Outcome, D = Treatment (Binary), Z = Instrument (Binary)
    dml_data = dml.DoubleMLData(df_g, 
                                y_col='percentile', 
                                d_cols='small_class', 
                                z_cols='initial_small', 
                                x_cols=x_cols)
    
    # Define Machine Learning algorithms (Random Forests)
    # ml_g: Predicts Y given X and Z
    ml_g = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
    # ml_m: Predicts Z given X (Propensity score of instrument)
    ml_m = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
    # ml_r: Predicts D given X and Z (Treatment probability)
    ml_r = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
    
    # Set numpy random seed for reproducible sample splitting in DoubleML
    np.random.seed(42)
    
    # Initialize Interactive IV Model (IIVM) for LATE estimation (removed random_state)
    dml_iivm = dml.DoubleMLIIVM(dml_data, ml_g, ml_m, ml_r, n_folds=5)
    
    # Fit the model
    dml_iivm.fit()
    
    coef = dml_iivm.coef[0]
    se = dml_iivm.se[0]
    
    print(f"Grade {grade} - DDML LATE: {coef:.3f} ({se:.3f})")
    return coef, se

# Run for all grades:
for g in [0, 1, 2, 3]:
    calculate_late_ddml(long_df, g)
```

## IPSW LATE

```python
import statsmodels.api as sm

def calculate_late_ipsw(data, grade):
    df_g = data[data['grade'] == grade].copy()
    
    covariates = [
        'white_asian', 'girl', 'free_lunch', 
        'teacher_white', 'teacher_exp', 'teacher_masters'
    ]
    school_dummies = pd.get_dummies(df_g['school_id'].astype(str), prefix='sch', drop_first=True)
    df_g = pd.concat([df_g, school_dummies], axis=1)
    x_cols = covariates + list(school_dummies.columns)
    
    df_g = df_g.dropna(subset=['percentile', 'small_class', 'initial_small'] + covariates)
    
    Y = df_g['percentile'].values
    T = df_g['small_class'].values
    Z = df_g['initial_small'].values
    X = sm.add_constant(df_g[x_cols].astype(float))
    
    # 1. Estimate Instrument Propensity Score E[Z|X] using Logit
    logit_model = sm.Logit(Z, X).fit(disp=0)
    p_X = logit_model.predict(X)
    
    # Trim probabilities to avoid division by zero / extreme weights
    p_X = np.clip(p_X, 0.01, 0.99)
    
    # 2. Calculate Normalized Weights (Hajek Estimator)
    w1 = Z / p_X
    w0 = (1 - Z) / (1 - p_X)
    
    w1_norm = w1 / np.sum(w1)
    w0_norm = w0 / np.sum(w0)
    
    # 3. Calculate Numerator (Average Causal Effect of Z on Y)
    itt_Y = np.sum(w1_norm * Y) - np.sum(w0_norm * Y)
    
    # 4. Calculate Denominator (Average Causal Effect of Z on T)
    itt_T = np.sum(w1_norm * T) - np.sum(w0_norm * T)
    
    # 5. IPSW LATE/ACR Estimate
    beta_ipsw = itt_Y / itt_T
    
    print(f"Grade {grade} - IPSW LATE: {beta_ipsw:.3f}")
    
    # Note: Analytical standard errors for IPSW require GMM or bootstrapping.
    return beta_ipsw

# Run for all grades:
for g in [0, 1, 2, 3]:
    calculate_late_ipsw(long_df, g)
```

## IPSW LATE SE

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

def compute_ipsw_point_estimate(df_est, x_cols):
    """Core function to calculate the IPSW LATE point estimate."""
    Y = df_est['percentile'].values
    T = df_est['small_class'].values
    Z = df_est['initial_small'].values
    X = sm.add_constant(df_est[x_cols].astype(float))
    
    # Estimate Instrument Propensity Score E[Z|X]
    logit_model = sm.Logit(Z, X).fit(disp=0, method='bfgs') 
    p_X = logit_model.predict(X)
    p_X = np.clip(p_X, 0.01, 0.99) # Trim to prevent extreme weights
    
    # Calculate Hajek Weights
    w1 = Z / p_X
    w0 = (1 - Z) / (1 - p_X)
    w1_norm = w1 / np.sum(w1)
    w0_norm = w0 / np.sum(w0)
    
    # Calculate components and ratio
    itt_Y = np.sum(w1_norm * Y) - np.sum(w0_norm * Y)
    itt_T = np.sum(w1_norm * T) - np.sum(w0_norm * T)

    logit_model = sm.Logit(Z, X).fit(disp=0, method='bfgs', warn_convergence=False)
    
    return itt_Y / itt_T

def calculate_late_ipsw_with_se(data, grade, B=500):
    """Calculates IPSW LATE and its cluster-robust standard error."""
    df_g = data[data['grade'] == grade].copy()
    
    covariates = [
        'white_asian', 'girl', 'free_lunch', 
        'teacher_white', 'teacher_exp', 'teacher_masters'
    ]
    school_dummies = pd.get_dummies(df_g['school_id'].astype(str), prefix='sch', drop_first=True)
    df_g = pd.concat([df_g, school_dummies], axis=1)
    x_cols = covariates + list(school_dummies.columns)
    
    df_g = df_g.dropna(subset=['percentile', 'small_class', 'initial_small', 'school_id'] + covariates)
    
    # 1. Calculate Point Estimate on full sample
    point_est = compute_ipsw_point_estimate(df_g, x_cols)
    
    # 2. Cluster Bootstrap for Standard Error
    clusters = df_g['school_id'].unique()
    n_clusters = len(clusters)
    boot_estimates = []
    
    np.random.seed(42) # Ensure reproducibility
    
    print(f"Grade {grade} - Running {B} bootstrap iterations...")
    for b in range(B):
        # Sample clusters with replacement
        boot_cluster_sample = np.random.choice(clusters, size=n_clusters, replace=True)
        
        # Reconstruct dataframe from sampled clusters
        boot_df_list = [df_g[df_g['school_id'] == cluster] for cluster in boot_cluster_sample]
        boot_df = pd.concat(boot_df_list, axis=0)
        
        try:
            boot_est = compute_ipsw_point_estimate(boot_df, x_cols)
            boot_estimates.append(boot_est)
        except Exception:
            # Logit may fail to converge in some bootstrap samples due to perfect separation
            continue
            
    # Compute Standard Error from the bootstrap distribution
    se_boot = np.std(boot_estimates, ddof=1)
    
    print(f"Grade {grade} - IPSW LATE: {point_est:.3f} ({se_boot:.3f})")
    print(f"Failed bootstrap iterations: {B - len(boot_estimates)}")
    
    return point_est, se_boot

# Run for a specific grade (e.g., Kindergarten)
ipsw_coef, ipsw_se = calculate_late_ipsw_with_se(long_df, grade=0, B=500)
```
